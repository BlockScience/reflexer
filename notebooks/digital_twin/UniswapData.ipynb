```python

# import libraries
import pandas as pd
import json
import requests
from tqdm.auto import tqdm
```

```python
QUERY = '''
{{
  uniswapSyncs (first: 1000, skip:{}) {{
    createdAt
    reserve0 # ETH balance
    reserve1 # RAI balance
  }}

  uniswapSwaps (first: 1000, skip:{}) {{
    createdAt 
    amount0In # ETH being locked
    amount1In # RAI being locked
    amount0Out # ETH being unlocked
    amount1Out # RAI being unlocked
  }}
}}
'''

url = 'https://api.thegraph.com/subgraphs/name/reflexer-labs/rai-mainnet'

pbar = tqdm()
uniswap_swaps = []
uniswap_syncs = []
i = 0
while True:
    query = QUERY.format(i, i)
    r = requests.post(url, json = {'query': query})
    parsed_json = json.loads(r.content)
    uniswap_swaps += parsed_json['data']['uniswapSwaps']
    uniswap_syncs += parsed_json['data']['uniswapSyncs']
    i += 1
    pbar.update(1)
    if i > 10:
        break
    
```
